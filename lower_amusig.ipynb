```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('mf_data.xlsx')
df.head()
```

```python
df_=df.copy()

plt.figure(figsize=(15, 6))  # Adjust width and height as needed
plt.plot(df_['datetime'], df_['lower_amusig'])
plt.xlabel('Datetime')
plt.ylabel('lower_amusig')
plt.title('lower_amusig Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Replace zeros in 'guihean' where the value is greater than 2 or less than 0.
df_.loc[(df_['lower_amusig'] > 1.2) | (df_['lower_amusig'] < 0), 'lower_amusig'] = 0
df_['lower_amusig'] = df_['lower_amusig'].replace(0, pd.NA).ffill()
```

```python
plt.figure(figsize=(15, 6))  # Adjust width and height as needed
plt.plot(df_['datetime'], df_['lower_amusig'])
plt.xlabel('Datetime')
plt.ylabel('lower_amusig')
plt.title('upper_amusig Over Time')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
df_mf_upper_amusig = df_[['datetime', 'lower_amusig']]
df_mf_upper_amusig.to_excel('df_mf_lower_amusig.xlsx')
```
